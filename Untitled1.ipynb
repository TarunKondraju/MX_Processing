```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, FFMpegWriter, PillowWriter
from pathlib import Path
import matplotlib as mpl
import imageio_ffmpeg

# Use the bundled ffmpeg from imageio-ffmpeg
mpl.rcParams["animation.ffmpeg_path"] = imageio_ffmpeg.get_ffmpeg_exe()

# ---------------------- CONFIG ----------------------
# Training data (demo)
np.random.seed(42)             # reproducible noise
X = np.atleast_2d(np.linspace(0, 10, 10)).T
y = np.sin(X).ravel()
dy = 0.2
y += dy * np.random.randn(y.shape[0])

# Prediction grid (keep moderate for speed/memory)
x_pred = np.atleast_2d(np.linspace(0, 10, 400)).T

# Gaussian Process
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C
kernel = C(1.0) * RBF(length_scale=1.0)

# Title schedule (time-based in seconds)
TITLE_SCHEDULE = [
    (0.0, "Step 1: PROSAIL-simulated + Sentinel-2 spectral inputs"),
    (3.0, "Step 2: Add LAI ground truth + EBD active learning"),
    (6.0, "Step 3: Optimized GPR model — uncertainty narrows"),
]

def get_title_for_time(t_seconds: float) -> str:
    current = TITLE_SCHEDULE[0][1]
    for t0, text in TITLE_SCHEDULE:
        if t_seconds >= t0:
            current = text
        else:
            break
    return current

# Output paths
out_dir = Path(r"D:\Tarun\iari\paper\GEE_application_paper")
out_dir.mkdir(parents=True, exist_ok=True)
mp4_path = out_dir / "gpr_ebd_lai_animation_4k.mp4"
gif_path = out_dir / "gpr_ebd_lai_animation_1080p_slow.gif"

# Playback speeds
FPS_MP4 = 12          # smooth for video
FPS_GIF = 1           # slow for readability (1 frame per second)

# Optional: add "holds" (duplicate frames) in the GIF to pause at steps
# e.g., hold at frame 2 for 2 extra seconds, and at frame 5 for 2 extra seconds
GIF_HOLDS = {2: 2, 5: 2}   # {frame_index: extra_seconds_to_hold}

# ---------------------- FIGURE (common) ----------------------
def make_figure(figsize_inches):
    fig, ax = plt.subplots(figsize=figsize_inches)
    line, = ax.plot([], [], label="GPR prediction", lw=2)
    pts,  = ax.plot([], [], "ko", label="Training data", markersize=6)

    # Title text object (updated per frame)
    title = ax.text(
        0.5, 1.05, "", ha="center", va="bottom",
        transform=ax.transAxes, fontsize=26, weight="bold"
    )

    ax.set_xlim(0, 10); ax.set_ylim(-2, 2)
    ax.set_xlabel("Sentinel-2 Spectral Inputs (Simulated + Observed)", fontsize=22)
    ax.set_ylabel("Estimated LAI", fontsize=22)
    ax.tick_params(axis='both', labelsize=20)
    ax.legend(loc="upper left", fontsize=18)
    return fig, ax, line, pts, title

# ---------------------- ANIMATION LOGIC ----------------------
def build_anim(fig, ax, line, pts, title, fps):
    band = {'poly': None}  # store in dict so closure can modify

    def init():
        title.set_text("")
        return line, pts, title

    def animate(i):
        # Fit GP with first i+1 training points
        gp = GaussianProcessRegressor(kernel=kernel, alpha=dy**2, n_restarts_optimizer=3)
        gp.fit(X[:i+1], y[:i+1])
        y_pred, s = gp.predict(x_pred, return_std=True)

        line.set_data(x_pred.ravel(), y_pred)
        pts.set_data(X[:i+1], y[:i+1])

        # Replace uncertainty band
        if band['poly'] is not None:
            band['poly'].remove()
        band['poly'] = ax.fill_between(
            x_pred.ravel(), y_pred - 1.96*s, y_pred + 1.96*s, alpha=0.2
        )

        # Time-based title
        t = i / float(fps)
        title.set_text(get_title_for_time(t))

        return line, pts, band['poly'], title

    frames = len(X)  # one frame per added training point
    anim = FuncAnimation(fig, animate, init_func=init, frames=frames, blit=False, repeat=False)
    return anim

# ---------------------- SAVE 4K MP4 ----------------------
# 4K: 19.2 x 10.8 inches at dpi=200 => ~3840x2160
fig4k, ax4k, line4k, pts4k, title4k = make_figure((19.2, 10.8))
ani4k = build_anim(fig4k, ax4k, line4k, pts4k, title4k, FPS_MP4)

writer = FFMpegWriter(
    fps=FPS_MP4,
    bitrate=6000,
    codec="h264",
    extra_args=["-pix_fmt", "yuv420p"]
)
ani4k.save(str(mp4_path), writer=writer, dpi=200)   # ~4K
print("Saved MP4:", mp4_path)

# ---------------------- SAVE 1080p GIF (slow + holds) ----------------------
# 1080p: 19.2 x 10.8 inches at dpi=100 => ~1920x1080
figgif, axgif, linegif, ptnsgif, titlegif = make_figure((19.2, 10.8))
anigif = build_anim(figgif, axgif, linegif, ptnsgif, titlegif, FPS_GIF)

# To create pauses in a GIF, we duplicate frames.
# Build a list of frame indices with duplicates for holds.
frame_indices = []
for i in range(len(X)):
    frame_indices.append(i)
    if i in GIF_HOLDS:
        # duplicate frame i, GIF_HOLDS[i] times (each duplication adds 1 second at FPS_GIF=1)
        frame_indices.extend([i] * GIF_HOLDS[i])

# Recreate animation with custom frames
def animate_with_indices(j):
    # j is an index into frame_indices; map to actual frame i
    i = frame_indices[j]
    # Reuse the inner logic by calling the same 'animate' as in build_anim.
    # We reimplement minimally here for clarity:
    gp = GaussianProcessRegressor(kernel=kernel, alpha=dy**2, n_restarts_optimizer=3)
    gp.fit(X[:i+1], y[:i+1])
    y_pred, s = gp.predict(x_pred, return_std=True)

    linegif.set_data(x_pred.ravel(), y_pred)
    ptnsgif.set_data(X[:i+1], y[:i+1])

    # Replace band
    # Find existing PolyCollections; remove the last band if present
    # (safer than keeping a handle when we rebuild anim)
    for coll in list(axgif.collections):
        coll.remove()
    axgif.fill_between(x_pred.ravel(), y_pred - 1.96*s, y_pred + 1.96*s, alpha=0.2)

    # Time-based title using FPS_GIF
    t = i / float(FPS_GIF)
    titlegif.set_text(get_title_for_time(t))
    return linegif, ptnsgif, titlegif

anigif = FuncAnimation(
    figgif, animate_with_indices, frames=len(frame_indices), blit=False, repeat=False
)

# Slower GIF via low FPS; pauses from duplicated frames.
anigif.save(str(gif_path), writer=PillowWriter(fps=FPS_GIF), dpi=100)
print("Saved GIF:", gif_path)
```

```python
# -*- coding: utf-8 -*-
"""
Recreate an artistic vertical infographic:
Machine ("GPR + EBD Active Learning") -> two outputs with separate arrows
Left: LAI Estimated Map (map icon)
Right: LAI Estimation Model (chip icon)
Transparent background; saves PNG (4K) + SVG.
"""

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch, Rectangle, Circle, Polygon
from matplotlib import patheffects as pe

# ---------- Canvas ----------
dpi = 200                     # 19.2x10.8 in @ 200 dpi ≈ 3840x2160 (4K)
fig = plt.figure(figsize=(19.2, 10.8), dpi=dpi, facecolor=(1,1,1,0))  # transparent bg
ax = plt.axes([0, 0, 1, 1])
ax.set_xlim(0, 100); ax.set_ylim(0, 100); ax.axis("off")

# ---------- Helpers ----------
def add_text(x, y, s, size=28, weight="bold", color="#0f172a", ha="center", va="center"):
    t = ax.text(x, y, s, fontsize=size, weight=weight, color=color, ha=ha, va=va)
    t.set_path_effects([pe.withStroke(linewidth=5, foreground="white", alpha=0.75)])
    return t

def rounded_box(x, y, w, h, fc, ec, lw=3, r=12):
    box = FancyBboxPatch((x, y), w, h, boxstyle=f"round,pad=0.8,rounding_size={r}",
                         linewidth=lw, edgecolor=ec, facecolor=fc)
    ax.add_patch(box)
    return box

def arrow(p1, p2, color="#f59e0b", lw=6, rad=0.0, head=24):
    arr = FancyArrowPatch(p1, p2, arrowstyle='-|>', mutation_scale=head,
                          linewidth=lw, color=color,
                          connectionstyle=f"arc3,rad={rad}")
    ax.add_patch(arr)
    return arr

def draw_gear(cx, cy, r=6, teeth=8, color="#6b21a8", ec="#2e1065", lw=2.5):
    # simple radial gear
    pts = []
    for i in range(teeth*2):
        ang = i*np.pi/teeth
        rr = r*1.0 if i%2==0 else r*0.72
        pts.append([cx + rr*np.cos(ang), cy + rr*np.sin(ang)])
    gear = Polygon(pts, closed=True, facecolor=color, edgecolor=ec, linewidth=lw)
    ax.add_patch(gear)
    ax.add_patch(Circle((cx, cy), r*0.32, facecolor="#111827", edgecolor=ec, linewidth=lw))
    return gear

def draw_machine():
    # main body
    body = rounded_box(26, 55, 48, 27, fc="#fb923c", ec="#1f2937", lw=4, r=10)
    # side window
    rounded_box(66, 60, 6, 10, fc="#0ea5e9", ec="#1f2937", lw=3, r=3)
    # indicator lights
    for i, c in enumerate(["#16a34a", "#f59e0b", "#ef4444"]):
        ax.add_patch(Circle((41 + i*4.8, 73.5), 1.2, facecolor=c, edgecolor="#1f2937", linewidth=2))
    # display plate
    plate = rounded_box(42, 66.5, 10, 5, fc="#60a5fa", ec="#1f2937", lw=3, r=2)
    # gears
    draw_gear(42, 63, r=6.5)
    draw_gear(52.5, 60.5, r=4.8)
    # top label
    header = rounded_box(42, 83.5, 16, 6, fc="#67e8f9", ec="#0f766e", lw=4, r=4)
    add_text(50, 86.5, "GPR + EBD\nActive Learning", size=28, weight="bold", color="#0f172a")
    return body

def draw_map_icon(x, y, w=20, h=14):
    # rounded container
    cont = rounded_box(x, y, w, h, fc="#0ea5e9", ec="#0f172a", lw=4, r=4)
    # grid tiles - stylized LAI map
    rng = np.random.default_rng(2)
    cols = ["#14532d","#166534","#16a34a","#22c55e","#65a30d","#a3e635","#f59e0b","#ef4444"]
    for i in range(5):
        for j in range(7):
            cx = x + 1.2 + j*(w-2.4)/7.2
            cy = y + 1.0 + i*(h-2.0)/5.2
            cw = (w-3.0)/7.0 * rng.uniform(0.85, 1.05)
            ch = (h-2.0)/5.2 * rng.uniform(0.85, 1.05)
            col = cols[(i*7+j) % len(cols)]
            ax.add_patch(Rectangle((cx, cy), cw, ch, facecolor=col, edgecolor="none", alpha=0.95))
    return cont

def draw_chip_icon(x, y, w=20, h=12, label="LAI"):
    # chip body
    chip = rounded_box(x, y, w, h, fc="#fde68a", ec="#0f172a", lw=4, r=2.5)
    # pins
    for i in range(6):
        ax.add_patch(Rectangle((x-1.8, y+1.5+i*(h-3)/5), 1.8, (h-3)/7, facecolor="#0f172a"))
        ax.add_patch(Rectangle((x+w, y+1.5+i*(h-3)/5), 1.8, (h-3)/7, facecolor="#0f172a"))
    add_text(x+w/2, y+h/2, label, size=26, weight="bold", color="#0f172a")
    return chip

# ---------- Compose ----------
# Machine
machine = draw_machine()

# Two independent arrows down-left and down-right from machine bottom
arrow((50, 55), (35, 41), color="#f59e0b", lw=8, rad=0.15, head=30)  # to left output
arrow((50, 55), (65, 41), color="#f59e0b", lw=8, rad=-0.15, head=30) # to right output

# LEFT output: LAI Estimated Map (map icon)
draw_map_icon(20, 25, w=23, h=16)
add_text(31.5, 20.5, "LAI Estimated Map", size=26, weight="bold", color="#111827")

# RIGHT output: LAI Estimation Model (chip icon)
draw_chip_icon(57, 26.5, w=22, h=13, label="LAI")
add_text(68, 20.5, "LAI Estimation Model", size=26, weight="bold", color="#111827")

# ---------- Save ----------
png_path = "gpr_ebd_vertical_outputs_4k.png"
svg_path = "gpr_ebd_vertical_outputs.svg"
fig.savefig(png_path, dpi=dpi, bbox_inches="tight", transparent=True)
fig.savefig(svg_path, bbox_inches="tight", transparent=True)  # vector-editable
print("Saved:", png_path, "and", svg_path)
```
