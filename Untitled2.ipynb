```python
# ==== Try-hard soil moisture modeling (aiming to maximize R²) ====
import math, json, joblib, numpy as np, pandas as pd, matplotlib.pyplot as plt
from pathlib import Path
from sklearn.model_selection import train_test_split, KFold, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler, PolynomialFeatures, QuantileTransformer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.ensemble import ExtraTreesRegressor, HistGradientBoostingRegressor, StackingRegressor
from sklearn.linear_model import Ridge
from sklearn.compose import TransformedTargetRegressor

# ---------- 0) Load ----------
csv_path = Path("D:\\Tarun\\iari\\Presentations\\Soil_moisture_GEE\\soil_moisture_vv_vh_engineered.csv")  # adjust if needed
df = pd.read_csv(csv_path)

# infer likely columns if present
candidate_y = [c for c in df.columns if c.lower() in ("soil_moisture","sm","target")]
assert len(candidate_y) > 0, "Couldn't find soil moisture column!"
y_col = candidate_y[0]

base_feats = [c for c in df.columns if c.upper() in ("VV","VH","ANGLE")]
# common engineered names (use any that exist)
more = [c for c in ["VV_sq","VH_sq","VVxVH","ratio","diff"] if c in df.columns]
all_feats = list(dict.fromkeys(base_feats + more))

X = df[all_feats].copy()
y = df[y_col].astype(float).copy()

# ---------- 1) Feature engineering ----------
# If angle exists, add trig transforms. Guard with presence.
if "ANGLE" in [c.upper() for c in X.columns]:
    angle_name = [c for c in X.columns if c.upper()=="ANGLE"][0]
    # radian transform
    ang = np.deg2rad(X[angle_name].astype(float))
    X["angle_rad"] = ang
    X["sin_angle"] = np.sin(ang)
    X["cos_angle"] = np.cos(ang)
    all_feats += ["angle_rad","sin_angle","cos_angle"]

# interaction terms (VV*angle, VH*angle) when available
if "VV" in X and "angle_rad" in X:
    X["VVxAngle"] = X["VV"] * X["angle_rad"]
    all_feats += ["VVxAngle"]
if "VH" in X and "angle_rad" in X:
    X["VHxAngle"] = X["VH"] * X["angle_rad"]
    all_feats += ["VHxAngle"]

# simple winsorization to reduce outlier influence (optional)
for c in X.columns:
    lo, hi = X[c].quantile(0.01), X[c].quantile(0.99)
    X[c] = X[c].clip(lo, hi)

# ---------- 2) Train/Validation split ----------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

# ---------- 3) Models ----------
# ExtraTrees: strong on tabular with many interactions
et = ExtraTreesRegressor(
    n_estimators=1200,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    max_features="sqrt",
    bootstrap=False,
    random_state=42,
    n_jobs=-1
)

# HistGradientBoosting: very competitive and fast
hgb = HistGradientBoostingRegressor(
    learning_rate=0.05,
    max_depth=None,
    max_leaf_nodes=64,
    min_samples_leaf=10,
    l2_regularization=1.0,
    random_state=42
)

# --- your blender pipeline ---
poly = PolynomialFeatures(degree=2, include_bias=False)
scaler = StandardScaler()

# 👇 Force SVD solver to avoid SciPy 'sym_pos' call
ridge = Ridge(alpha=1.0, solver="svd", random_state=42)

linear_blender = Pipeline([
    ("poly", poly),
    ("scaler", scaler),
    ("ridge", ridge),
])

# keep your target transform
lin_T = TransformedTargetRegressor(
    regressor=linear_blender,
    transformer=QuantileTransformer(output_distribution="normal", random_state=42)
)

# # Light polynomial ridge “blender” on scaled features (helps capture smooth trends)
# linear_blender = Pipeline([
#     ("poly", PolynomialFeatures(degree=2, include_bias=False)),
#     ("scaler", StandardScaler()),
#     ("ridge", Ridge(alpha=1.0, random_state=42))
# ])

# Randomized search spaces (kept sensible to finish in minutes)
et_search = RandomizedSearchCV(
    estimator=et,
    param_distributions={
        "n_estimators": [600, 900, 1200, 1500],
        "max_features": ["sqrt","log2", None],
        "min_samples_leaf": [1, 2, 4, 8],
        "min_samples_split": [2, 5, 10]
    },
    n_iter=25, cv=KFold(5, shuffle=True, random_state=42),
    scoring="r2", n_jobs=-1, random_state=42, verbose=0
)

hgb_search = RandomizedSearchCV(
    estimator=hgb,
    param_distributions={
        "learning_rate": [0.02, 0.03, 0.05, 0.08],
        "max_leaf_nodes": [31, 63, 127],
        "min_samples_leaf": [5, 10, 20, 30],
        "l2_regularization": [0.0, 0.1, 0.5, 1.0, 2.0]
    },
    n_iter=25, cv=KFold(5, shuffle=True, random_state=42),
    scoring="r2", n_jobs=-1, random_state=42, verbose=0
)

# Target transform often boosts R² for skewed moisture distributions
qtx = QuantileTransformer(output_distribution="normal", random_state=42)
et_T = TransformedTargetRegressor(regressor=et_search, transformer=qtx)
hgb_T = TransformedTargetRegressor(regressor=hgb_search, transformer=qtx)
lin_T = TransformedTargetRegressor(regressor=linear_blender, transformer=qtx)

# Fit base models (with internal CV search)
print("Fitting ExtraTrees (with RandomizedSearchCV)...")
et_T.fit(X_train, y_train)
print("Best ET params:", et_T.regressor_.best_params_)

print("Fitting HistGradientBoosting (with RandomizedSearchCV)...")
hgb_T.fit(X_train, y_train)
print("Best HGB params:", hgb_T.regressor_.best_params_)

print("Fitting polynomial Ridge blender...")
lin_T.fit(X_train, y_train)

# Stacking ensemble (meta-model is a small Ridge on standardized base predictions)
stack = StackingRegressor(
    estimators=[
        ("et", et_T),
        ("hgb", hgb_T),
        ("lin", lin_T)
    ],
    final_estimator=Ridge(alpha=0.5, random_state=42),
    passthrough=False,
    n_jobs=-1
)

print("Fitting Stacking ensemble...")
stack.fit(X_train, y_train)

# ---------- 4) Evaluate ----------
def eval_and_report(model, name):
    y_pred = model.predict(X_test)
    r2 = r2_score(y_test, y_pred)
    rmse = math.sqrt(mean_squared_error(y_test, y_pred))
    mae = mean_absolute_error(y_test, y_pred)
    nrmse = rmse / (y_test.max() - y_test.min())
    sd = np.std(y_test, ddof=1)
    rpd = sd / rmse if rmse != 0 else np.inf
    print(f"{name:>14s} | R²={r2:.3f}  RMSE={rmse:.3f}  MAE={mae:.3f}  NRMSE={nrmse:.3f}  RPD={rpd:.3f}")
    return y_pred, {"name": name, "r2": r2, "rmse": rmse, "mae": mae, "nrmse": nrmse, "rpd": rpd}

preds = {}
metrics = []
for mdl, name in [(et_T,"ExtraTrees(T)"), (hgb_T,"HGB(T)"), (lin_T,"PolyRidge(T)"), (stack,"STACK")]:
    yp, m = eval_and_report(mdl, name)
    preds[name] = yp
    metrics.append(m)

# pick best by R²
best = max(metrics, key=lambda d: d["r2"])
best_name = best["name"]
print("\nBest model:", best_name, "metrics:", best)

# ---------- 5) Plot Predicted vs Measured (best), with 1:1 + trendline ----------
y_pred_best = preds[best_name]
fig, ax = plt.subplots(figsize=(7,6))
ax.scatter(y_test, y_pred_best, s=30, alpha=0.7, edgecolor="none")

# 1:1 line
mn = min(y_test.min(), y_pred_best.min())
mx = max(y_test.max(), y_pred_best.max())
ax.plot([mn, mx], [mn, mx], lw=2)

# trend line
coef = np.polyfit(y_test, y_pred_best, 1)
xl = np.linspace(mn, mx, 100)
ax.plot(xl, coef[0]*xl + coef[1], lw=2, linestyle="--")

ax.set_xlabel("Measured SM")
ax.set_ylabel("Predicted SM")
ax.set_title("RF/Ensemble Estimated vs Measured SM")

# metrics box (kept outside to avoid overlap)
txt = (f"{best_name}\n"
       f"R² = {best['r2']:.3f}\n"
       f"RMSE = {best['rmse']:.3f}\n"
       f"MAE = {best['mae']:.3f}\n"
       f"RPD = {best['rpd']:.3f}")
ax.legend([ "1:1 line", f"Trend: y={coef[0]:.2f}x+{coef[1]:.2f}" ],
          loc="upper left", bbox_to_anchor=(0,1.15))
ax.text(1.02, 0.5, txt, transform=ax.transAxes,
        verticalalignment="center", bbox=dict(boxstyle="round", alpha=0.15))

plt.tight_layout()
out_png = Path("pred_vs_meas_best.png")
plt.savefig(out_png, dpi=200)
plt.close()

# ---------- 6) Save best model & info ----------
# store the actual fitted best object
name_to_model = {"ExtraTrees(T)": et_T, "HGB(T)": hgb_T, "PolyRidge(T)": lin_T, "STACK": stack}
best_model = name_to_model[best_name]
joblib.dump(best_model, "best_soil_moisture_model.joblib")
with open("best_metrics.json","w") as f:
    json.dump(best, f, indent=2)
with open("features_used.json","w") as f:
    json.dump(all_feats, f, indent=2)

print("\nSaved:")
print("  - best_soil_moisture_model.joblib")
print("  - best_metrics.json")
print("  - features_used.json")
print(f"  - Pred/Meas plot: {out_png}")
```

```python
#MLR recalculation

import argparse, sys, math
from dataclasses import dataclass
from typing import List, Dict, Tuple
import numpy as np, pandas as pd
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
from joblib import dump

try:
    from statsmodels.stats.outliers_influence import variance_inflation_factor
    _HAS_STATS = True
except Exception:
    _HAS_STATS = False

@dataclass
class Metrics:
    r2: float
    rmse: float
    nrmse: float
    rpd: float

def compute_metrics(y_true, y_pred) -> Metrics:
    y_true = np.asarray(y_true).ravel()
    y_pred = np.asarray(y_pred).ravel()
    r2 = r2_score(y_true, y_pred)
    rmse = mean_squared_error(y_true, y_pred, squared=False)
    ymin, ymax = float(np.min(y_true)), float(np.max(y_true))
    nrmse = rmse / (ymax - ymin) if ymax > ymin else np.nan
    rpd = (np.std(y_true, ddof=1) / rmse) if rmse > 0 else np.nan
    return Metrics(r2, rmse, nrmse, rpd)

def add_engineered_features(df: pd.DataFrame, vv: str, vh: str) -> pd.DataFrame:
    out = df.copy()
    eps = 1e-12
    out['VV_sq'] = out[vv] ** 2
    out['VH_sq'] = out[vh] ** 2
    out['VVxVH'] = out[vv] * out[vh]
    out['diff'] = out[vh] - out[vv]
    out['ratio'] = out[vh] / (out[vv] + eps)
    return out

def db_to_linear_cols(df: pd.DataFrame, cols: List[str]) -> pd.DataFrame:
    out = df.copy()
    for c in cols:
        out[c] = np.power(10.0, out[c] / 10.0)
    return out

def compute_vif(df_X: pd.DataFrame) -> pd.Series:
    if not _HAS_STATS:
        raise RuntimeError("statsmodels not available for VIF computation.")
    X = df_X.values.astype(float)
    vals = []
    for i in range(X.shape[1]):
        try: vals.append(variance_inflation_factor(X, i))
        except Exception: vals.append(np.inf)
    return pd.Series(vals, index=df_X.columns, dtype=float)

def vif_prune(df_X: pd.DataFrame, vif_threshold: float = 10.0, max_rounds: int = 10):
    if vif_threshold <= 0: return df_X, {}
    dropped = {}
    X = df_X.copy()
    rounds = 0
    while rounds < max_rounds and X.shape[1] > 1:
        try: vifs = compute_vif(X)
        except RuntimeError: return df_X, {}
        max_v = vifs.max()
        if np.isfinite(max_v) and max_v <= vif_threshold: break
        worst = vifs.idxmax()
        dropped[worst] = float(vifs.loc[worst])
        X = X.drop(columns=[worst])
        rounds += 1
    return X, dropped

def run_kfold_mlr(df, feature_cols, target_col, n_splits=5, vif_threshold=10.0,
                  use_ridge_if_singular=True, ridge_alpha=1e-6):
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
    y = df[target_col].values.astype(float)
    X_full = df[feature_cols].copy()
    metrics_list, fold_diags = [], {}

    for fold_idx, (tr, te) in enumerate(kf.split(X_full, y)):
        X_tr_raw, y_tr = X_full.iloc[tr].copy(), y[tr]
        X_te_raw, y_te = X_full.iloc[te].copy(), y[te]

        X_tr_pruned, dropped = vif_prune(X_tr_raw, vif_threshold=vif_threshold)
        kept_cols = list(X_tr_pruned.columns)

        scaler = StandardScaler().fit(X_tr_pruned.values.astype(float))
        X_tr = scaler.transform(X_tr_pruned.values.astype(float))
        X_te = scaler.transform(X_te_raw[kept_cols].values.astype(float))

        model = LinearRegression()
        try:
            model.fit(X_tr, y_tr)
        except Exception:
            if use_ridge_if_singular:
                model = Ridge(alpha=ridge_alpha, random_state=42).fit(X_tr, y_tr)
            else:
                raise

        y_pred = model.predict(X_te)
        metrics_list.append(compute_metrics(y_te, y_pred))
        fold_diags[fold_idx] = {"kept_features": kept_cols, "dropped_by_vif": dropped,
                                "n_train": int(len(tr)), "n_test": int(len(te))}

    # Final fit on all data (VIF prune once more)
    X_all_pruned, dropped_all = vif_prune(X_full, vif_threshold=vif_threshold)
    kept_all = list(X_all_pruned.columns)
    scaler_all = StandardScaler().fit(X_all_pruned.values.astype(float))
    X_all = scaler_all.transform(X_all_pruned.values.astype(float))
    final_model = LinearRegression()
    try:
        final_model.fit(X_all, y)
    except Exception:
        if use_ridge_if_singular:
            final_model = Ridge(alpha=ridge_alpha, random_state=42).fit(X_all, y)
        else:
            raise

    coef_df = pd.DataFrame({"feature": kept_all,
                            "coef_standardized": final_model.coef_.ravel().tolist()})
    coef_df.loc[len(coef_df)] = ["<intercept_std_space>", float(final_model.intercept_)]
    return metrics_list, coef_df, {"final_kept_features": kept_all,
                                   "final_dropped_by_vif": dropped_all}

def run_kfold_rf(df, feature_cols, target_col, n_splits=5, param_grid=None):
    if param_grid is None:
        param_grid = [
            {"n_estimators": 300, "min_samples_leaf": 1},
            {"n_estimators": 300, "min_samples_leaf": 4},
            {"n_estimators": 500, "min_samples_leaf": 1},
            {"n_estimators": 500, "min_samples_leaf": 4},
            {"n_estimators": 1000, "min_samples_leaf": 1},
            {"n_estimators": 1000, "min_samples_leaf": 4},
        ]
    y = df[target_col].values.astype(float)
    X = df[feature_cols].values.astype(float)
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)

    grid_results = []
    for params in param_grid:
        fold_ms = []
        for tr, te in kf.split(X, y):
            model = RandomForestRegressor(n_estimators=params["n_estimators"],
                                          min_samples_leaf=params["min_samples_leaf"],
                                          random_state=42, n_jobs=-1)
            model.fit(X[tr], y[tr])
            pred = model.predict(X[te])
            fold_ms.append(compute_metrics(y[te], pred))
        mean_r2 = float(np.mean([m.r2 for m in fold_ms]))
        mean_rmse = float(np.mean([m.rmse for m in fold_ms]))
        grid_results.append({"params": params, "mean_r2": mean_r2, "mean_rmse": mean_rmse})

    best = sorted(grid_results, key=lambda d: (-d["mean_r2"], d["mean_rmse"]))[0]
    best_params = best["params"]

    fold_metrics, fold_diags = [], {}
    for i, (tr, te) in enumerate(kf.split(X, y)):
        model = RandomForestRegressor(n_estimators=best_params["n_estimators"],
                                      min_samples_leaf=best_params["min_samples_leaf"],
                                      random_state=42, n_jobs=-1)
        model.fit(X[tr], y[tr])
        pred = model.predict(X[te])
        fold_metrics.append(compute_metrics(y[te], pred))
        fold_diags[i] = {"n_train": int(len(tr)), "n_test": int(len(te))}
    final_model = RandomForestRegressor(**best_params, random_state=42, n_jobs=-1).fit(X, y)
    return fold_metrics, fold_diags, {"best_params": best_params, "final_model": final_model}

def summarize_metrics(name: str, per_fold: List[Metrics]) -> pd.DataFrame:
    dfm = pd.DataFrame([m.__dict__ for m in per_fold])
    mean_row = dfm.mean(numeric_only=True)
    summary = pd.DataFrame([mean_row]); summary.index = [f"{name}_MEAN"]
    return dfm, summary

def main():
    p = argparse.ArgumentParser("Soil moisture modeling: MLR (engineered features + VIF) vs RF")
    p.add_argument("--csv", required=True)
    p.add_argument("--target", default="soil_moisture")
    p.add_argument("--vv", default="VV")
    p.add_argument("--vh", default="VH")
    p.add_argument("--theta", default=None)
    p.add_argument("--db_to_linear", action="store_true")
    p.add_argument("--vif_threshold", type=float, default=10.0)
    p.add_argument("--kfold", type=int, default=5)
    p.add_argument("--save_models", action="store_true")
    args = p.parse_args()

    df = pd.read_csv(args.csv)
    for c in [args.vv, args.vh, args.target]:
        if c not in df.columns:
            sys.exit(f"Missing column '{c}'. Found: {list(df.columns)}")
    if args.theta and args.theta not in df.columns:
        sys.exit(f"Missing theta column '{args.theta}'.")

    if args.db_to_linear:
        df = db_to_linear_cols(df, [args.vv, args.vh])

    df = add_engineered_features(df, args.vv, args.vh)

    feature_cols = [args.vv, args.vh, "VV_sq", "VH_sq", "VVxVH", "diff", "ratio"]
    if args.theta: feature_cols.append(args.theta)

    use_cols = feature_cols + [args.target]
    df = df.dropna(subset=use_cols).reset_index(drop=True)
    print("Data shape after dropna:", df.shape)
    print("Features:", feature_cols)

    # MLR
    mlr_folds, mlr_coefs, mlr_diags = run_kfold_mlr(
        df, feature_cols, args.target,
        n_splits=args.kfold, vif_threshold=args.vif_threshold
    )
    mlr_folds_df, mlr_mean_df = summarize_metrics("MLR", mlr_folds)
    print("\nMLR per-fold:\n", mlr_folds_df.round(4))
    print("\nMLR mean:\n", mlr_mean_df.round(4))
    print("\nMLR kept (final):", mlr_diags["final_kept_features"])
    print("MLR dropped by VIF (final):", mlr_diags["final_dropped_by_vif"])
    print("\nFinal MLR coefficients (standardized space):\n", mlr_coefs)

    # RF (same predictors)
    rf_folds, rf_diags, rf_info = run_kfold_rf(df, feature_cols, args.target, n_splits=args.kfold)
    rf_folds_df, rf_mean_df = summarize_metrics("RF", rf_folds)
    print("\nRF per-fold:\n", rf_folds_df.round(4))
    print("\nRF mean:\n", rf_mean_df.round(4))
    print("\nRF best params:", rf_info["best_params"])

    if args.save_models:
        dump(rf_info["final_model"], "rf_final.joblib")
        print("Saved RF -> rf_final.joblib")

    mlr_folds_df.to_csv("mlr_cv_folds.csv", index=False)
    mlr_mean_df.to_csv("mlr_cv_mean.csv")
    rf_folds_df.to_csv("rf_cv_folds.csv", index=False)
    rf_mean_df.to_csv("rf_cv_mean.csv")
    print("Saved: mlr_cv_folds.csv, mlr_cv_mean.csv, rf_cv_folds.csv, rf_cv_mean.csv")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd, numpy as np
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
from statsmodels.stats.outliers_influence import variance_inflation_factor

# ----------------- SETTINGS -----------------
CSV_PATH = r"D:\Tarun\iari\Soil_Moisture\soil_moisture_vv_vh_engineered.csv"
TARGET = "soil_moisture"
VV = "VV"
VH = "VH"
THETA_CANDIDATES = ["localIncidenceAngle_mean", "localIncidenceAngle_mean.1"]
DB_TO_LINEAR = False  # set True to convert dB -> linear power
K = 5
VIF_THRESHOLD = 10.0  # set 0 to disable pruning
# --------------------------------------------

def compute_metrics(y_true, y_pred):
    y_true = np.asarray(y_true).ravel()
    y_pred = np.asarray(y_pred).ravel()
    r2 = r2_score(y_true, y_pred)
    rmse = mean_squared_error(y_true, y_pred, squared=False)
    nrmse = rmse / (np.max(y_true) - np.min(y_true))
    rpd = np.std(y_true, ddof=1) / rmse if rmse > 0 else np.nan
    return {"R2": r2, "RMSE": rmse, "NRMSE": nrmse, "RPD": rpd}

def add_engineered(df, vv, vh):
    out = df.copy()
    eps = 1e-12
    out["VV_sq"] = out[vv] ** 2
    out["VH_sq"] = out[vh] ** 2
    out["VVxVH"] = out[vv] * out[vh]
    out["diff"] = out[vh] - out[vv]
    out["ratio"] = out[vh] / (out[vv] + eps)
    return out

def pick_theta(df, candidates):
    # choose the candidate with non-zero variance; prefer the first that varies
    for c in candidates:
        if c in df.columns:
            if np.isfinite(df[c]).all() and df[c].std(ddof=1) > 1e-12:
                return c
    return None

def drop_zero_variance(dfX):
    keep = [c for c in dfX.columns if dfX[c].std(ddof=1) > 1e-12]
    dropped = [c for c in dfX.columns if c not in keep]
    return dfX[keep].copy(), dropped

def vif_prune(dfX, threshold=10.0, max_rounds=10):
    if threshold <= 0 or dfX.shape[1] <= 1:
        return dfX.copy(), {}
    dropped = {}
    X = dfX.copy()
    rounds = 0
    while rounds < max_rounds and X.shape[1] > 1:
        vifs = []
        for i in range(X.shape[1]):
            try:
                vifs.append(variance_inflation_factor(X.values, i))
            except Exception:
                vifs.append(np.inf)
        vifs = pd.Series(vifs, index=X.columns, dtype=float)
        max_v = float(np.nanmax(vifs.values))
        if not np.isfinite(max_v) or max_v <= threshold:
            break
        worst = vifs.idxmax()
        dropped[worst] = max_v
        X = X.drop(columns=[worst])
        rounds += 1
    return X, dropped

# -------- Load, prep, and diagnose --------
df = pd.read_csv(CSV_PATH)

# Choose theta if viable
theta = pick_theta(df, THETA_CANDIDATES)
print("Chosen theta column:", theta)

# dB -> linear (optional)
df2 = df.copy()
if DB_TO_LINEAR:
    for c in [VV, VH]:
        df2[c] = 10 ** (df2[c] / 10.0)

# Build engineered set
df2 = add_engineered(df2, VV, VH)

# Candidate feature list (identical for both models)
feature_cols = [VV, VH, "VV_sq", "VH_sq", "VVxVH", "diff", "ratio"]
if theta is not None:
    feature_cols.append(theta)

# Clean NaNs
df2 = df2.dropna(subset=feature_cols + [TARGET]).reset_index(drop=True)

# Drop zero-variance features
X0 = df2[feature_cols].copy()
X1, dropped_zero_var = drop_zero_variance(X0)
print("Dropped zero-variance:", dropped_zero_var)

# VIF prune
X2, dropped_vif = vif_prune(X1, threshold=VIF_THRESHOLD)
print("Dropped by VIF:", dropped_vif)
kept = list(X2.columns)
print("Final kept features:", kept)

# --------- MLR (with VIF-pruned features) ---------
kf = KFold(n_splits=K, shuffle=True, random_state=42)
mlr_metrics = []
for tr, te in kf.split(df2):
    X_tr, X_te = df2.iloc[tr][kept].values, df2.iloc[te][kept].values
    y_tr, y_te = df2.iloc[tr][TARGET].values, df2.iloc[te][TARGET].values
    scaler = StandardScaler().fit(X_tr)
    X_tr_s, X_te_s = scaler.transform(X_tr), scaler.transform(X_te)
    model = LinearRegression()
    try:
        model.fit(X_tr_s, y_tr)
    except Exception:
        model = Ridge(alpha=1e-6, random_state=42).fit(X_tr_s, y_tr)
    pred = model.predict(X_te_s)
    mlr_metrics.append(compute_metrics(y_te, pred))
mlr_mean = pd.DataFrame(mlr_metrics).mean().to_frame("MLR_mean")
print(mlr_mean.round(4))

# --------- RF (same cleaned features) ----------
rf = RandomForestRegressor(n_estimators=500, min_samples_leaf=1, random_state=42, n_jobs=-1)
rf_metrics = []
for tr, te in kf.split(df2):
    X_tr, X_te = df2.iloc[tr][kept].values, df2.iloc[te][kept].values
    y_tr, y_te = df2.iloc[tr][TARGET].values, df2.iloc[te][TARGET].values
    rf.fit(X_tr, y_tr)
    pred = rf.predict(X_te)
    rf_metrics.append(compute_metrics(y_te, pred))
rf_mean = pd.DataFrame(rf_metrics).mean().to_frame("RF_mean")
print(rf_mean.round(4))
```

```python
# === Best-index and Best-subset MLR finder (Notebook-friendly) ===
# Paste this entire cell and run.

import itertools
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# ------------------ USER SETTINGS ------------------
CSV_PATH = r"D:\Tarun\iari\Soil_Moisture\soil_moisture_vv_vh_engineered.csv"  # change if needed
TARGET   = "soil_moisture"
VV_COL   = "VV"
VH_COL   = "VH"
THETA_CANDIDATES = ["localIncidenceAngle_mean", "localIncidenceAngle_mean.1"]  # we’ll auto-pick a usable one

DB_TO_LINEAR = False   # set True to convert VV/VH from dB -> linear power
KFOLD_SPLITS = 5       # CV folds
MAX_COMBO_SIZE = 3     # test up to 3-variable subsets
TOPK_PLOTS = 3         # how many top single-index plots to draw
# ---------------------------------------------------

def add_engineered_features(df, vv, vh):
    out = df.copy()
    eps = 1e-12
    out["VV_sq"]  = out[vv] ** 2
    out["VH_sq"]  = out[vh] ** 2
    out["VVxVH"]  = out[vv] * out[vh]
    out["diff"]   = out[vh] - out[vv]
    out["ratio"]  = out[vh] / (out[vv] + eps)
    return out

def choose_theta(df, candidates):
    for c in candidates:
        if c in df.columns:
            s = df[c]
            if np.isfinite(s).all() and s.std(ddof=1) > 1e-12:
                return c
    return None

def compute_metrics(y_true, y_pred):
    y_true = np.asarray(y_true).ravel()
    y_pred = np.asarray(y_pred).ravel()
    r2 = r2_score(y_true, y_pred)
    rmse = mean_squared_error(y_true, y_pred, squared=False)
    rng = np.max(y_true) - np.min(y_true)
    nrmse = rmse / rng if rng > 0 else np.nan
    rpd = (np.std(y_true, ddof=1) / rmse) if rmse > 0 else np.nan
    return {"R2": r2, "RMSE": rmse, "NRMSE": nrmse, "RPD": rpd}

def cv_r2_for_columns(df, feature_cols, target_col, k_splits=5):
    """Return mean CV R2 for a given set of feature columns (handles singular designs gracefully)."""
    X = df[feature_cols].values
    y = df[target_col].values
    kf = KFold(n_splits=k_splits, shuffle=True, random_state=42)
    r2s = []
    for tr, te in kf.split(X, y):
        Xtr, Xte = X[tr], X[te]
        ytr, yte = y[tr], y[te]
        try:
            mdl = LinearRegression()
            mdl.fit(Xtr, ytr)
            pred = mdl.predict(Xte)
            r2s.append(r2_score(yte, pred))
        except Exception:
            # Singular matrix or other numerical issue -> treat as very poor
            r2s.append(float("nan"))
    r2s = np.array(r2s, dtype=float)
    return np.nanmean(r2s)

def fit_equation(df, feature_cols, target_col):
    """Fit OLS on all data and return coefficients, intercept, in-sample R2, RMSE."""
    X = df[feature_cols].values
    y = df[target_col].values
    mdl = LinearRegression().fit(X, y)
    pred = mdl.predict(X)
    ins_r2 = r2_score(y, pred)
    ins_rmse = mean_squared_error(y, pred, squared=False)
    return mdl.coef_, mdl.intercept_, ins_r2, ins_rmse

# ------------------ LOAD & PREP ------------------
df = pd.read_csv(CSV_PATH)

# Convert dB to linear if requested
df2 = df.copy()
if DB_TO_LINEAR:
    for c in [VV_COL, VH_COL]:
        df2[c] = np.power(10.0, df2[c] / 10.0)

# Add indices
df2 = add_engineered_features(df2, VV_COL, VH_COL)

# Pick a usable incidence angle column
theta_col = choose_theta(df2, THETA_CANDIDATES)

# Build candidate feature set
base_features = [VV_COL, VH_COL, "VV_sq", "VH_sq", "VVxVH", "diff", "ratio"]
if theta_col:
    base_features.append(theta_col)

# Drop rows with NaNs in used columns
use_cols = base_features + [TARGET]
df2 = df2.dropna(subset=use_cols).reset_index(drop=True)

print(f"Data rows after cleaning: {len(df2)}")
print("Candidate predictors:", base_features)

# ------------------ 1) BEST SINGLE INDEX ------------------
single_results = []
for col in base_features:
    mean_r2_cv = cv_r2_for_columns(df2, [col], TARGET, KFOLD_SPLITS)
    coef, intercept, ins_r2, ins_rmse = fit_equation(df2, [col], TARGET)
    single_results.append({
        "Index": col,
        "CV_R2": mean_r2_cv,
        "FullFit_Coeff": float(coef[0]),
        "FullFit_Intercept": float(intercept),
        "FullFit_R2_in_sample": float(ins_r2),
        "FullFit_RMSE_in_sample": float(ins_rmse),
    })

single_df = pd.DataFrame(single_results).sort_values("CV_R2", ascending=False)
print("\n=== Best single-index models (sorted by CV_R2) ===")
display(single_df)

# ------------------ 2) BEST SUBSET (2- and 3-variable) ------------------
subset_rows = []
max_k = min(MAX_COMBO_SIZE, len(base_features))
for r in range(2, max_k + 1):
    for combo in itertools.combinations(base_features, r):
        combo = list(combo)
        mean_r2_cv = cv_r2_for_columns(df2, combo, TARGET, KFOLD_SPLITS)
        # Fit full-data equation for reporting:
        try:
            coef, intercept, ins_r2, ins_rmse = fit_equation(df2, combo, TARGET)
            coef_map = {name: float(val) for name, val in zip(combo, coef)}
        except Exception:
            coef_map = {name: float("nan") for name in combo}
            intercept, ins_r2, ins_rmse = float("nan"), float("nan"), float("nan")
        subset_rows.append({
            "Predictors": tuple(combo),
            "CV_R2": mean_r2_cv,
            "FullFit_Intercept": float(intercept),
            "FullFit_R2_in_sample": float(ins_r2),
            "FullFit_RMSE_in_sample": float(ins_rmse),
            **{f"coef_{k}": v for k, v in coef_map.items()}
        })

subset_df = pd.DataFrame(subset_rows).sort_values("CV_R2", ascending=False)
print("\n=== Best multi-variable subsets (2..N, sorted by CV_R2) ===")
display(subset_df.head(15))

# ------------------ 3) PRINT TOP EQUATIONS ------------------
def pretty_eq(coefs, intercept, names):
    terms = [f"{co:.6g}·{nm}" for co, nm in zip(coefs, names)]
    return f"Soil_moisture = " + " + ".join(terms) + f" + {intercept:.6g}"

print("\n=== TOP SINGLE-INDEX EQUATIONS ===")
for i, row in single_df.head(5).iterrows():
    idx = row["Index"]
    coef = row["FullFit_Coeff"]
    intercept = row["FullFit_Intercept"]
    eq = pretty_eq([coef], intercept, [idx])
    print(f"[{idx}]  CV_R2={row['CV_R2']:.4f} | FullFit_R2={row['FullFit_R2_in_sample']:.4f} | {eq}")

print("\n=== TOP MULTI-VARIABLE EQUATIONS (first 5) ===")
for _, row in subset_df.head(5).iterrows():
    names = [k.replace("coef_", "") for k in row.index if k.startswith("coef_")]
    coefs = [row[f"coef_{n}"] for n in names]
    eq = pretty_eq(coefs, row["FullFit_Intercept"], names)
    print(f"{tuple(names)}  CV_R2={row['CV_R2']:.4f} | FullFit_R2={row['FullFit_R2_in_sample']:.4f} | {eq}")

# ------------------ 4) PLOTS: top-K single-index fits ------------------
topk = min(TOPK_PLOTS, len(single_df))
for rank in range(topk):
    row = single_df.iloc[rank]
    idx = row["Index"]
    # Fit full model (one predictor)
    X = df2[[idx]].values
    y = df2[TARGET].values
    mdl = LinearRegression().fit(X, y)
    x_plot = np.linspace(X.min(), X.max(), 200).reshape(-1, 1)
    y_plot = mdl.predict(x_plot)

    plt.figure(figsize=(6, 4))
    plt.scatter(X, y, s=22)
    plt.plot(x_plot, y_plot, linewidth=2)
    plt.xlabel(idx)
    plt.ylabel(TARGET)
    plt.title(f"{idx} → {TARGET} | R² (CV)={row['CV_R2']:.3f}")
    plt.tight_layout()
    plt.show()

# ------------------ 5) SAVE SUMMARIES ------------------
single_df.to_csv("best_single_index_MLR.csv", index=False)
subset_df.to_csv("best_subset_MLR.csv", index=False)
print("\nSaved: best_single_index_MLR.csv, best_subset_MLR.csv")
```
