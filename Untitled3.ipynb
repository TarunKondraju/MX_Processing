```python
from graphviz import Digraph

def vasudha_flowchart_4k(output_name="vasudha_flowchart_4k"):
    dot = Digraph(format="png")
    dot.attr(rankdir="TB", dpi="400")   # DPI adjusted for 4K resolution

    # Nodes
    dot.node("start", "Start", shape="oval")

    dot.node("sensor", "Choose Sensor\n(VNIR or MIR)", shape="box")
    dot.node("mode", "Choose Mode\n(Realtime or Offline)", shape="box")

    dot.node("rt", "Realtime Mode", shape="box")
    dot.node("offline", "Offline Mode", shape="box")

    dot.node("watch", "Select Watch Folder\n(Auto-read new spectra)", shape="box")

    dot.node("soil", "Select Soil Type\n(Alluvial, Red, Black, etc.)", shape="box")

    dot.node("load_rt", "Spectra Arrive in Watch Folder", shape="box")
    dot.node("load_off", "Load Spectral Files Manually", shape="box")

    dot.node("pre", "Preprocess Spectrum\n(SG Filter, Scaling, Imputation)", shape="box")

    dot.node("predict", "Predict Soil Fertility Parameters\n(N, P, K, OC, pH, EC)", shape="box")

    dot.node("export", "Download Predictions,\nSpectral Data, and Plots", shape="box")

    dot.node("end", "End", shape="oval")

    # Edges
    dot.edge("start", "sensor")
    dot.edge("sensor", "mode")

    dot.edge("mode", "rt", label="Realtime")
    dot.edge("mode", "offline", label="Offline")

    # Realtime path
    dot.edge("rt", "watch")
    dot.edge("watch", "soil")
    dot.edge("soil", "load_rt")
    dot.edge("load_rt", "pre")

    # Offline path
    dot.edge("offline", "soil")
    dot.edge("soil", "load_off")
    dot.edge("load_off", "pre")

    dot.edge("pre", "predict")
    dot.edge("predict", "export")
    dot.edge("export", "end")

    # Render
    filename = dot.render(output_name, cleanup=True)
    print(f"Flowchart saved as: {filename}")

# Generate the flowchart
vasudha_flowchart_4k()
```

```python
# Artistic 4K Flowchart for VASUDHA (Matplotlib-only)
# Produces: VASUDHA_Artistic_Flowchart_4K.png
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch
from matplotlib import cm

# --- Canvas 4K ---
FIG_W, FIG_H = 38.4, 21.6   # inches at dpi=100 -> 3840x2160
DPI = 100
fig = plt.figure(figsize=(FIG_W, FIG_H), dpi=DPI)
ax = fig.add_axes([0,0,1,1])
ax.set_xlim(0, 100)
ax.set_ylim(-40, 100)
ax.axis('off')

# --- Background gradient ---
nx, ny = 400, 200
X = np.linspace(0,1,nx)
Y = np.linspace(0,1,ny)
G = np.outer(np.ones(ny), X)
# Use a soft teal-to-indigo gradient
c1 = np.array([0.11,0.55,0.55])   # teal
c2 = np.array([0.15,0.0,0.4])     # indigo
grad = np.zeros((ny, nx, 3))
for i in range(3):
    grad[:, :, i] = c1[i] * (1-G) + c2[i] * (G)
ax.imshow(grad, extent=[0,100,-40,100], origin='lower', alpha=0.95)

# subtle vignette (darken edges)
xx = np.linspace(-1,1, nx)
yy = np.linspace(-1,1, ny)
XX, YY = np.meshgrid(xx, yy)
V = np.exp(-2*(XX**2 + YY**2))
vign = np.dstack([V,V,V])
ax.imshow((1-0.12*(1-vign)), extent=[0,100,-40,100], origin='lower', alpha=0.3)

# --- Style helpers ---
def rounded_box(x, y, w, h, text, box_color='#ffffff', text_color='#082f3a',
                fontsize=26, shadow=True, alpha=0.98):
    # shadow
    if shadow:
        sh = FancyBboxPatch((x+0.8, y-0.8), w, h,
                            boxstyle="round,pad=0.8,rounding_size=8",
                            linewidth=0, facecolor='black', alpha=0.12, zorder=1)
        ax.add_patch(sh)
    box = FancyBboxPatch((x, y), w, h,
                         boxstyle="round,pad=0.8,rounding_size=8",
                         linewidth=1.6, facecolor=box_color, edgecolor='#0b3a3f', zorder=2, alpha=alpha)
    ax.add_patch(box)
    # glossy highlight
    hi = FancyBboxPatch((x+0.7, y+h*0.55), w-1.4, h*0.35,
                        boxstyle="round,pad=0.6,rounding_size=6",
                        linewidth=0, facecolor='white', alpha=0.12, zorder=3)
    ax.add_patch(hi)
    ax.text(x+w/2, y+h/2, text, ha='center', va='center', fontsize=fontsize, color=text_color, zorder=4, wrap=True)

def curved_arrow(xy_from, xy_to, color='#ffffff', lw=2.2, rad=0.3):
    a = FancyArrowPatch(xy_from, xy_to, arrowstyle='-|>', connectionstyle=f"arc3,rad={rad}",
                        color=color, linewidth=lw, mutation_scale=22, zorder=3)
    ax.add_patch(a)

# --- Title ---
ax.text(50, 96, "VASUDHA Workflow", ha='center', va='center',
        fontsize=44, fontweight='700', color='white', zorder=5)

# --- Boxes positions & sizes (artistic layout) ---
# Start
rounded_box(42, 80, 16, 8, "START", box_color='#f7f9fa', text_color='#0b3a3f', fontsize=26)

# Sensor & Mode
rounded_box(30, 64, 40, 10, "Choose Sensor (VNIR or MIR)\nThen choose Mode (Realtime or Offline)",
            box_color='#f0fbfa', text_color='#07323a', fontsize=22)

# Realtime branch (left)
rounded_box(6, 44, 28, 9, "Realtime Mode", box_color='#ffffff', fontsize=22)
rounded_box(6, 31, 28, 9, "Select Watch Folder\n(Auto-read spectra)", box_color='#eaf6f6', fontsize=20)
rounded_box(6, 18, 28, 9, "Select Soil Type\n(Alluvial, Red, Black, ...)", box_color='#f7fbff', fontsize=20)
rounded_box(6, 6, 28, 9, "Software Reads Spectra →\nPreprocess & Predict", box_color='#fff7ea', fontsize=20)

# Offline branch (right)
rounded_box(66, 44, 28, 9, "Offline Mode", box_color='#ffffff', fontsize=22)
rounded_box(66, 31, 28, 9, "Select Soil Type", box_color='#f7fbff', fontsize=20)
rounded_box(66, 18, 28, 9, "Load Spectral Files\n(Manual)", box_color='#eaf6f6', fontsize=20)
rounded_box(66, 6, 28, 9, "Preprocess & Predict", box_color='#fff7ea', fontsize=20)

# Output & End
rounded_box(30, -16, 40, 12, "Download Predictions, Spectral Data\nand Plots (CSV / PNG)", box_color='#ffffff', fontsize=22)
rounded_box(42, -32, 16, 8, "END", box_color='#f7f9fa', fontsize=22)

# --- Arrows (curved, artistic) ---
curved_arrow((50, 80), (50, 74), color='white', lw=2.6, rad=0.0)                # start -> choose sensor
curved_arrow((50, 64), (35, 54), color='white', lw=2.2, rad=-0.25)              # to realtime branch
curved_arrow((50, 64), (65, 54), color='white', lw=2.2, rad=0.25)               # to offline branch

# Realtime chain arrows
curved_arrow((20, 53), (20, 44.5), color='#f2f2f2', lw=2.0, rad=0.0)
curved_arrow((20, 40.5), (20, 32), color='#f2f2f2', lw=2.0, rad=0.0)
curved_arrow((20, 25), (20, 16), color='#f2f2f2', lw=2.0, rad=0.0)

# Offline chain arrows
curved_arrow((80, 53), (80, 44.5), color='#f2f2f2', lw=2.0, rad=0.0)
curved_arrow((80, 40.5), (80, 32), color='#f2f2f2', lw=2.0, rad=0.0)
curved_arrow((80, 25), (80, 16), color='#f2f2f2', lw=2.0, rad=0.0)

# Merge arrows to output
curved_arrow((20, 6), (44, -4), color='#ffffff', lw=2.2, rad=0.2)
curved_arrow((80, 6), (56, -4), color='#ffffff', lw=2.2, rad=-0.2)

# Output to end
curved_arrow((50, -4), (50, -10), color='#ffffff', lw=2.4, rad=0.0)
curved_arrow((50, -20), (50, -26), color='#ffffff', lw=2.4, rad=0.0)

# # small decorative icons (circles) to indicate processes
# ax.add_patch(patches.Circle((14, 10), 1.1, color='#ffbf69', alpha=0.95, zorder=4))
# ax.text(14, 10, "PL", ha='center', va='center', fontsize=10, zorder=5, color='#082f3a')

# ax.add_patch(patches.Circle((86, 10), 1.1, color='#69b3ff', alpha=0.95, zorder=4))
# ax.text(86, 10, "FT", ha='center', va='center', fontsize=10, zorder=5, color='#082f3a')

# footer small text
ax.text(50, -36, "VASUDHA — Visualise & Assess Soil Using Digital Hyperspectral Analytics", ha='center', va='center',
        fontsize=14, color='white', alpha=0.9)

# --- Save 4K PNG ---
outname = "VASUDHA_Artistic_Flowchart_4K.png"
plt.savefig(outname, dpi=DPI, bbox_inches='tight', facecolor='none')
plt.close(fig)
print("Saved:", outname)
```
